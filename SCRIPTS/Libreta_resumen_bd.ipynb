# Resumen de la base de datos

En esta libreta se hacen las siguientes dos cosas:

1. Documentar que tipo de informacion almacena cada tabla de la base de datos;
2. Seleccionar los campos o columnas de interes de cada tabla.

El objetivo de **(1)** es el de enteder que informacion se almacena en la base de datos; comprendiendo asi cada uno de los campos de cada una de las tablas. El objetivo de **(2)** es el seleccionar que campos de cada tabla son de nuestro interes. Esto con el objetivo de crear queries mas simples que ignoren dichos campos.

Para llevar a cabo **(2)** primero tendremos que haber definido el objetivo del analisis. La base de datos es lo suficientemente extensa para plantear muchos analisis puntuales, los cuales no necesitarian toda la informacion de la base. Esto se definira en el [READ.me](../README.md).

### Tabla Player_Attributes

La tabla Player tiene las siguientes columnas:

- id: Llave primaria de la tabla; identificador interno del snapshot de atributos. Tipo de dato entero.
- player_fifa_api_id: Llave foranea. Identificador del jugador en la API de FIFA/SoFIFA. Tipo de dato entero.
- player_api_id: Llave foranea. Identificador del jugador en la fuente principal del dataset (osea, la llave primaria en la API de donde se obtuvieron los datos de los partidos). Tipo de dato entero.
- date: Fecha en la que se registro ese snapshot de atributos del jugador. Tipo de dato string con formato fecha-hora.
- overall_rating: Calificacion general del jugador en FIFA para esa fecha. Tipo de dato entero.
- potential: Potencial estimado del jugador en FIFA para esa fecha. Tipo de dato entero.
- preferred_foot: Pie dominante o preferido del jugador (left o right). Tipo de dato string.
- attacking_work_rate: Tendencia o intensidad de trabajo ofensivo del jugador; conceptualmente es una categoria textual tipo low, medium, high, aunque la base tiene algunos valores sucios. Tipo de dato string.
- defensive_work_rate: Tendencia o intensidad de trabajo defensivo del jugador; conceptualmente es una categoria textual tipo low, medium, high, aunque la base tiene algunos valores sucios. Tipo de dato string.
- crossing: Puntaje FIFA de la capacidad del jugador para centrar el balón. Tipo de dato entero.
- finishing: Puntaje FIFA de la capacidad de definicion/remate del jugador. Tipo de dato entero.
- heading_accuracy: Puntaje FIFA de la precision en remates de cabeza. Tipo de dato entero.
- short_passing: Puntaje FIFA de la calidad y precision en pases cortos. Tipo de dato entero.
- volleys: Puntaje FIFA de la capacidad de rematar de volea. Tipo de dato entero.
- dribbling: Puntaje FIFA de la habilidad de drible/conduccion con balon. Tipo de dato entero.
- curve: Puntaje FIFA de la capacidad de darle efecto o curvatura al balon. Tipo de dato entero.
- free_kick_accuracy: Puntaje FIFA de la precision en tiros libres. Tipo de dato entero.
- long_passing: Puntaje FIFA de la calidad y precisión en pases largos. Tipo de dato entero.
- ball_control: Puntaje FIFA del control de balon del jugador. Tipo de dato entero.
- acceleration: Puntaje FIFA de la aceleracion del jugador. Tipo de dato entero.
- sprint_speed: Puntaje FIFA de la velocidad maxima de sprint del jugador. Tipo de dato entero.
- agility: Puntaje FIFA de la agilidad del jugador. Tipo de dato entero.
- reactions: Puntaje FIFA de la rapidez de reaccion del jugador ante jugadas. Tipo de dato entero.
- balance: Puntaje FIFA del equilibrio corporal del jugador. Tipo de dato entero.
- shot_power: Puntaje FIFA de la potencia de disparo. Tipo de dato entero.
- jumping: Puntaje FIFA de la capacidad de salto. Tipo de dato entero.
- stamina: Puntaje FIFA de la resistencia física del jugador. Tipo de dato entero.
- strength: Puntaje FIFA de la fuerza física del jugador. Tipo de dato entero.
- long_shots: Puntaje FIFA de la efectividad en disparos de larga distancia. Tipo de dato entero.
- aggression: Puntaje FIFA de la agresividad competitiva del jugador. Tipo de dato entero.
- interceptions: Puntaje FIFA de la capacidad de interceptar balones o pases. Tipo de dato entero.
- positioning: Puntaje FIFA de la colocacion ofensiva del jugador. Tipo de dato entero.
- vision: Puntaje FIFA de la visión de juego para detectar opciones de pase. Tipo de dato entero.
- penalties: Puntaje FIFA de la capacidad de ejecutar penales. Tipo de dato entero.
- marking: Puntaje FIFA de la capacidad de marcar rivales. Tipo de dato entero.
- standing_tackle: Puntaje FIFA de la efectividad para quitar el balon con entrada de pie. Tipo de dato entero.
- sliding_tackle: Puntaje FIFA de la efectividad para quitar el balon con barrida/deslizamiento. Tipo de dato entero.
- gk_diving: Puntaje FIFA de la capacidad del portero para lanzarse. Tipo de dato entero.
- gk_handling: Puntaje FIFA de la capacidad del portero para atrapar/controlar el balon con las manos. Tipo de dato entero.
- gk_kicking: Puntaje FIFA de la calidad del despeje o golpeo del portero. Tipo de dato entero.

### Tabla Player

- [x] Documentacion finalizada?

Descripcion de sus columnas:

- id: Llave primaria de la tabla; identificador interno del registro del jugador. Tipo de dato entero.
- player_api_id: Identificador del jugador en la fuente principal de partidos/alineaciones del dataset (osea, la llave primaria en la API de donde se obtuvieron los datos de los partidos); es la llave usada para enlazar con Player_Attributes y con Match. Tipo de dato entero.
- player_name: Nombre completo del jugador. Tipo de dato string.
- player_fifa_api_id: Identificador del jugador en la API de FIFA/SoFIFA. Tipo de dato entero.
- birthday: Fecha de nacimiento del jugador. Tipo de dato string con formato fecha-hora (aunque hay que estandarizar formato y limpiar la base).
- height: Altura del jugador; en la práctica almacena centimetros con decimales. Tipo de dato float.
- weight: Peso del jugador; el peso esta en libras. Tipo de dato entero.

### Tabla Match

- [x] Documentacion finalizada?

Descripcion de sus columnas:

- id: Llave primaria de la tabla; identificador interno del partido. Tipo de dato entero.
- country_id: Llave foránea que referencia a Country.id; indica el país de la liga en que se jugó el partido. Tipo de dato entero.
- league_id: Llave foránea que referencia a League.id; indica la liga del partido. Tipo de dato entero.
- season: Temporada del partido, usualmente con formato YYYY/YYYY. Tipo de dato string.
- stage: Jornada, ronda o etapa del partido dentro de la temporada. Tipo de dato entero.
- date: Fecha y hora del partido. Tipo de dato string con formato fecha-hora.
- match_api_id: Identificador del partido en la fuente original de partidos del dataset. Tipo de dato entero.
- home_team_api_id: Identificador del equipo local en la fuente principal del dataset; referencia a Team.team_api_id. Tipo de dato entero.
- away_team_api_id: Identificador del equipo visitante en la fuente principal del dataset; referencia a Team.team_api_id. Tipo de dato entero.
- home_team_goal: Goles anotados por el equipo local. Tipo de dato entero.
- away_team_goal: Goles anotados por el equipo visitante. Tipo de dato entero.
- home_player_1 a home_player_11: Identificadores API de los once titulares del equipo local; enlazan con Player.player_api_id. Tipo de dato entero.
- away_player_1 a away_player_11: Identificadores API de los once titulares del equipo visitante; enlazan con Player.player_api_id. Tipo de dato entero.
- home_player_X1 a home_player_X11: Coordenada X del jugador local correspondiente dentro del esquema/formación en cancha. Tipo de dato entero.
- away_player_X1 a away_player_X11: Coordenada X del jugador visitante correspondiente dentro del esquema/formación en cancha. Tipo de dato entero.
- home_player_Y1 a home_player_Y11: Coordenada Y del jugador local correspondiente dentro del esquema/formación en cancha. Tipo de dato entero.
- away_player_Y1 a away_player_Y11: Coordenada Y del jugador visitante correspondiente dentro del esquema/formación en cancha. Tipo de dato entero.
- goal: Registro en XML/texto con los eventos de gol del partido. Tipo de dato string.
- shoton: Registro en XML/texto con los tiros a puerta del partido. Tipo de dato string.
- shotoff: Registro en XML/texto con los tiros fuera del arco. Tipo de dato string.
- foulcommit: Registro en XML/texto con las faltas cometidas en el partido. Tipo de dato string.
- card: Registro en XML/texto con las tarjetas mostradas en el partido. Tipo de dato string.
- cross: Registro en XML/texto con eventos de centros. Tipo de dato string.
- corner: Registro en XML/texto con eventos de tiros de esquina. Tipo de dato string.
- possession: Registro en XML/texto con eventos o tramos de posesión del balón. Tipo de dato string.
- B365H, B365D, B365A: Cuotas de apuesta pre-partido de Bet365 para victoria local, empate y victoria visitante. Tipo de dato numerico.
- BWH, BWD, BWA: Cuotas de apuesta pre-partido de Bet&Win/Bwin para victoria local, empate y victoria visitante. Tipo de dato numerico.
- IWH, IWD, IWA: Cuotas de apuesta pre-partido de Interwetten para victoria local, empate y victoria visitante. Tipo de dato numerico.
- LBH, LBD, LBA: Cuotas de apuesta pre-partido de Ladbrokes para victoria local, empate y victoria visitante. Tipo de dato numerico.
- PSH, PSD, PSA: Cuotas de apuesta pre-partido de Pinnacle para victoria local, empate y victoria visitante. Tipo de dato numerico.
- WHH, WHD, WHA: Cuotas de apuesta pre-partido de William Hill para victoria local, empate y victoria visitante. Tipo de dato numerico.
- SJH, SJD, SJA: Cuotas de apuesta pre-partido de Stan James para victoria local, empate y victoria visitante. Tipo de dato numerico.
- VCH, VCD, VCA: Cuotas de apuesta pre-partido de VC Bet para victoria local, empate y victoria visitante. Tipo de dato numerico.
- GBH, GBD, GBA: Cuotas de apuesta pre-partido de Gamebookers para victoria local, empate y victoria visitante. Tipo de dato numerico.
- BSH, BSD, BSA: Cuotas de apuesta pre-partido de Blue Square para victoria local, empate y victoria visitante. Tipo de dato numerico.

### Tabla Team_Attributes

- [x] Documentacion finalizada?

Descripcion de sus columnas:

- id: Llave primaria de la tabla; identificador interno del snapshot tactico del equipo. Tipo de dato entero.
- team_fifa_api_id: Identificador del equipo en FIFA/SoFIFA. Tipo de dato entero.
- team_api_id: Identificador del equipo en la fuente principal de partidos del dataset; enlaza con Team en team_api_id. Tipo de dato entero.
- date: Fecha del snapshot tactico/estadistico del equipo en FIFA. Tipo de dato string con formato fecha-hora.
- buildUpPlaySpeed: Puntaje FIFA de la velocidad con la que el equipo construye jugadas. Tipo de dato entero.
- buildUpPlaySpeedClass: Clasificación cualitativa del ritmo de construccion del equipo. Tipo de dato string.
- buildUpPlayDribbling: Puntaje FIFA de cuanto usa el equipo la conduccion/drible durante la salida o construccion. Tipo de dato entero.
- buildUpPlayDribblingClass: Clasificacion cualitativa del uso de drible en la construccion. Tipo de dato string.
- buildUpPlayPassing: Puntaje FIFA de la tendencia del equipo a construir con pases mas directos/largos o mas elaborados. Tipo de dato entero.
- buildUpPlayPassingClass: Clasificacion cualitativa del estilo de pase en la construccion. Tipo de dato string.
- buildUpPlayPositioningClass: Tipo de posicionamiento del equipo en la construcción de juego, por ejemplo Organised o Free Form. Tipo de dato string.
- chanceCreationPassing: Puntaje FIFA de cuanto genera ocasiones por medio del pase. Tipo de dato entero.
- chanceCreationPassingClass: Clasificacion cualitativa de la creacion de ocasiones por pase. Tipo de dato string.
- chanceCreationCrossing: Puntaje FIFA de cuanto genera ocasiones por medio de centros. Tipo de dato entero.
- chanceCreationCrossingClass: Clasificacion cualitativa de la creación de ocasiones por centros. Tipo de dato string.
- chanceCreationShooting: Puntaje FIFA de la tendencia del equipo a terminar jugadas con disparos. Tipo de dato entero.
- chanceCreationShootingClass: Clasificacion cualitativa de la generacion de ocasiones mediante tiros. Tipo de dato string.
- chanceCreationPositioningClass: Tipo de posicionamiento ofensivo del equipo al generar ocasiones, por ejemplo Organised o Free Form. Tipo de dato string.
- defencePressure: Puntaje FIFA de la intensidad de presion defensiva del equipo. Tipo de dato entero.
- defencePressureClass: Clasificacion cualitativa del nivel de presion defensiva. Tipo de dato string.
- defenceAggression: Puntaje FIFA de la agresividad defensiva del equipo al presionar o entrar. Tipo de dato entero.
- defenceAggressionClass: Clasificacion cualitativa de la agresividad defensiva. Tipo de dato string.
- defenceTeamWidth: Puntaje FIFA de la amplitud con la que el equipo defiende. Tipo de dato entero.
- defenceTeamWidthClass: Clasificacion cualitativa de la anchura defensiva del equipo. Tipo de dato string.
- defenceDefenderLineClass: Comportamiento de la linea defensiva, por ejemplo Cover u Offside Trap. Tipo de dato string.
